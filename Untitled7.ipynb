```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
# Replace 'your_dataset.csv' with the actual path to your file
try:
    df = pd.read_csv('/content/Titanic-Dataset.csv')
except FileNotFoundError:
    print("Error: 'your_dataset.csv' not found. Please replace it with your data file.")
    # Create a sample DataFrame for demonstration if the file is not found
    data = {'col1': [1, 2, 3, np.nan, 5],
            'col2': ['A', 'B', 'C', 'D', 'E'],
            'col3': [10.1, 20.2, np.nan, 40.4, 50.5]}
    df = pd.DataFrame(data)

# Display the first few rows
print("First 5 rows of the DataFrame:")
display(df.head())

# Get information about the DataFrame (data types, non-null counts)
print("\nDataFrame Info:")
display(df.info())

# Get descriptive statistics
print("\nDescriptive Statistics:")
display(df.describe())

# Check for missing values
print("\nMissing values per column:")
display(df.isnull().sum())
# Example 1: Dropping rows with any missing values
# df_dropped_rows = df.dropna()

# Example 2: Dropping columns with any missing values
# df_dropped_cols = df.dropna(axis=1)

# Example 3: Imputing missing values with the mean of the column
# For numerical columns
for col in df.select_dtypes(include=np.number).columns:
    df[col].fillna(df[col].mean(), inplace=True)

# Example 4: Imputing missing values with the mode of the column
# For categorical columns
for col in df.select_dtypes(include='object').columns:
    df[col].fillna(df[col].mode()[0], inplace=True)

print("\nDataFrame after handling missing values:")
display(df.isnull().sum())
# Check for duplicate rows
print("\nNumber of duplicate rows:", df.duplicated().sum())

# Drop duplicate rows
df.drop_duplicates(inplace=True)

print("Number of duplicate rows after removal:", df.duplicated().sum())
# Example: Visualize numerical data distribution with box plots to identify outliers
for col in df.select_dtypes(include=np.number).columns:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col])
    plt.title(f'Box plot of {col}')
    plt.show()

# Example: Simple outlier removal using the Interquartile Range (IQR)
# This is a basic example, more sophisticated methods exist
for col in df.select_dtypes(include=np.number).columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

print("\nDataFrame shape after potential outlier removal:", df.shape)
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Example: Min-Max Scaling
scaler = MinMaxScaler()
numerical_cols = df.select_dtypes(include=np.number).columns
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

# Example: Standardization (commented out, choose one)
# scaler = StandardScaler()
# df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

print("\nDataFrame after feature scaling:")
display(df.head())
# Example: One-Hot Encoding
df = pd.get_dummies(df, columns=df.select_dtypes(include='object').columns, drop_first=True)

print("\nDataFrame after one-hot encoding:")
display(df.head())
# Example: Creating a new feature by combining two existing features
# Assuming 'feature1' and 'feature2' are numerical columns
# if 'feature1' in df.columns and 'feature2' in df.columns:
#     df['new_feature'] = df['feature1'] * df['feature2']

# Example: Extracting information from a date column
# Assuming 'date_col' is a datetime column (you might need to convert it first)
# if 'date_col' in df.columns:
#     df['year'] = pd.to_datetime(df['date_col']).dt.year
#     df['month'] = pd.to_datetime(df['date_col']).dt.month
#     df['day'] = pd.to_datetime(df['date_col']).dt.day

print("\nDataFrame after potential feature engineering:")
display(df.head())

```
